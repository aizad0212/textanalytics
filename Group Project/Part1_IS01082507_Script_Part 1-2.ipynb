```python
# 3.Text Pre-Processing

# import library 
import re 
import string 
import nltk 
from nltk.corpus import stopwords
from textblob import TextBlob
import emoji
```

```python
stop_words = stopwords.words('english') 
more_stopwords = ['u', 'im', 'c'] 
stop_words = stop_words + more_stopwords
```

```python
stemmer = nltk.SnowballStemmer("english")
```

```python
internet_slang = {
    "u": "you",
    "ur": "your",
    "pls": "please",
    "plz": "please",
    "thx": "thanks",
    "btw": "by the way",
    "idk": "i do not know",
    "imo": "in my opinion",
    "lmao": "laughing my ass off",
    "lol": "laughing out loud",
    "omg": "oh my god",
    "gr8": "great",
    "b4": "before",
    "cuz": "because",
    "cu": "see you",
    "tnx": "thanks",
    "tq": "thank you",
    "nvm": "never mind",
    "wanna": "want to",
    "gonna": "going to",
    "gotta": "got to",
    "ya": "yeah",
    "nah": "no",
    "ok": "okay",
    "kk": "okay",
    "bc": "because",
    "r": "are",
    "ya'll": "you all"
}

contractions = {
    "i'm": "i am",
    "you're": "you are",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "we're": "we are",
    "they're": "they are",
    "i've": "i have",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "i'd": "i would",
    "you'd": "you would",
    "he'd": "he would",
    "she'd": "she would",
    "we'd": "we would",
    "they'd": "they would",
    "i'll": "i will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "we'll": "we will",
    "they'll": "they will",
    "can't": "cannot",
    "won't": "will not",
    "don't": "do not",
    "doesn't": "does not",
    "didn't": "did not",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "hasn't": "has not",
    "haven't": "have not",
    "hadn't": "had not",
    "shouldn't": "should not",
    "wouldn't": "would not",
    "couldn't": "could not",
    "mustn't": "must not",
    "mightn't": "might not",
    "needn't": "need not"
}
```

```python
def preprocess(text):
    # Lowercase
    text = text.lower()

    # Remove expressions and content in square brackets
    text = re.sub(r'\[.*?\]', '', text)

    # Remove URLs
    text = re.sub(r'http\S+\s*\S+', '', text)
    text = re.sub(r'www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove emojis
    text = emoji.replace_emoji(text, replace='')

    # Replace internet slang
    for slang, replacement in internet_slang.items():
        text = re.sub(r'\b' + re.escape(slang) + r'\b', replacement, text)

    # Replace contractions
    for contraction, full in contractions.items():
        text = re.sub(r'\b' + re.escape(contraction) + r'\b', full, text)

    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Remove numbers
    text = re.sub(r'\b\w*\d\w*\b', '', text)

    # Correct spelling mistakes using TextBlob
    text = str(TextBlob(text).correct())

    # Remove stopwords
    text = ' '.join(word for word in text.split() if word not in stop_words)

    # Stemming
    text = ' '.join(stemmer.stem(word) for word in text.split())

    return text
```

```python
df_cleaned['Cleaned_Reviews'] = df_cleaned['Reviews'].apply(preprocess) 
df_cleaned.head(20)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer and stemmer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

# Define the stop words
stop_words = set(stopwords.words('english'))
```

```python
# Function to perform stemming and lemmatization
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Tokenize the text into words and convert to lowercase
    tokens = [token for token in tokens if token.isalnum()]  # Filter out non-alphanumeric tokens
    tokens = [token for token in tokens if token not in stop_words]  # Remove stopwords from the tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatize each token
    return tokens  # Return the preprocessed tokens
```

```python
# Apply lemmatization and stemming
df_cleaned['Processed_Reviews'] = df_cleaned['Cleaned_Reviews'].apply(preprocess_text)  # Apply the function to each document

# Display the first 20 rows of the cleaned data
df_cleaned.head(20)
```
